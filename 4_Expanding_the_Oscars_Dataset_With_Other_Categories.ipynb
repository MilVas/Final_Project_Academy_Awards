### Creating an empty dataframe to add a list of all directors and all years combined

```python
ceremonies=all_oscars.ceremony.tolist()
ceremonies = set(ceremonies)
ceremonies1 = list(ceremonies)


directors = best_directing_cat.name.tolist()
directors=set(directors)
directors =list(directors)


bestpic_directors = pictures_combined3.Director.tolist()
bestpic_directors = set(bestpic_directors)
bestpic_directors = list(bestpic_directors)


all_directors = directors+bestpic_directors
all_directors=set(all_directors)
all_directors = list(all_directors)
```

```python
from itertools import product

df_directors  = pd.DataFrame(list(product(ceremonies, all_directors)), columns=['ceremony', 'director'])
```

```python
df_directors
```

```python
# splitting first and second director
df_directors[['first_dir','second_dir']] = df_directors.director.str.split(',', expand = True)
#getting rid of unecessary columns
df_directors = df_directors.drop(columns=['director','second_dir'])
df_directors
```

```python
# adding best directing nominations

df_directors1 = pd.merge(df_directors,best_directing_cat,
                        how='left',
                        left_on=['first_dir','ceremony'],
                        right_on=['name','ceremony'])
```

```python
df_directors1
```

```python
df_directors1 = df_directors1.rename(columns={'winner':'best_directing_win'})
df_directors1.best_directing_win = df_directors1.best_directing_win.replace({True:1,False:0})
```

```python
df_directors1 = df_directors1.drop(columns=['name'])
```

```python
#dropping duplicates and unecessary columns

df_directors1  = df_directors1.drop_duplicates(subset=['ceremony', 'first_dir'], keep='first')
```

```python
# replacing NaN values with 0
df_directors1=df_directors1.fillna(0)
```

```python
# Calculating cumulative sums for nominations and wins of directors
```

```python
noms = df_directors1.copy()
```

```python
noms1 = noms.set_index(['first_dir','ceremony'])
noms2 = noms1.groupby(level=[0]).cumsum()
noms2 =noms2.reset_index()
```

```python
noms2 = noms2.rename(columns={'nominated_for_best_director':"best_directing_nomination_sum",
                             'best_directing_win':'best_directing_win_sum'})
```

```python
noms3 = noms2.sort_values(by=['first_dir','ceremony'])
```

```python
noms3 = noms3.set_index(['first_dir','ceremony'])
```

```python
noms4 = noms3.groupby(level=[0]).diff()
noms4 = noms4.reset_index()
noms4 = noms4.fillna(0)
```

```python
#merging sum of nominations/wins with the difference in order to get prior nominations and wins

directors_final_list = pd.merge(noms2, noms4,
                               how='left',
                               on=['first_dir','ceremony'])
```

```python
directors_final_list['prior_best_directing_nominations'] = (directors_final_list['best_directing_nomination_sum_x']
                                                            -directors_final_list['best_directing_nomination_sum_y'])
                                                            
                                                            
directors_final_list['prior_best_directing_wins'] = (directors_final_list['best_directing_win_sum_x']
                                                            -directors_final_list['best_directing_win_sum_y'])                                                    
```

```python
directors_final_list=directors_final_list.rename(columns={'best_directing_nomination_sum_x':'best_directing_nomination_sum',
                                                         'best_directing_win_sum_x':'best_directing_win_sum'})


directors_final_list = directors_final_list.drop(columns=['best_directing_nomination_sum_y','best_directing_win_sum_y'])
```

```python
#merging directors wiht the best picture dataset

pictures_combined3 = pd.merge(pictures_combined3,directors_final_list,
                             how='left',
                             left_on=['Director','ceremony'],
                             right_on=['first_dir','ceremony'])
```

```python
pictures_combined3
```

```python
#adding a category to calculate the number of films the director had nominated for 'best picture' category
```

```python
total_best_pic = pictures_combined3[['Director','ceremony']]
total_best_pic['moviez'] = 1
total_best_pic = total_best_pic.sort_values(['Director','ceremony'])
```

```python
total_best_pic = total_best_pic.groupby(['Director','ceremony'])['moviez'].count().reset_index()
total_best_pic1 = total_best_pic.set_index(['Director','ceremony'])
total_best_pic1=total_best_pic1.groupby(level=[0]).cumsum()
total_best_pic1=total_best_pic1.reset_index()
total_best_pic1 = total_best_pic1.rename(columns={'moviez':'running_best_picture_nomination'})
```

```python
pictures_combined4 = pd.merge(pictures_combined3,total_best_pic1,
                             how='left',
                             on=['Director','ceremony']
                             )
```

# 2  Golden Globes

```python
#import previuos version of Oscars dataset. 
pictures_combined4 = pd.read_csv('data/pictures_combined4.csv')

#import dataset from Kaggle
golden_globes = pd.read_csv('data/Golden Globe awards dataset kaggle/golden_globe_awards_K.csv')
```

```python
# renaming 1917 in oscars dataset
pictures_combined4.film = pictures_combined4.film.str.replace('1917;','1917')

#renaming category'winner' to 'Best_Pic_Winner'
pictures_combined4 = pictures_combined4.rename(columns={'winner':'Best_Pic_Winner'})


#stripping potential blank spaces
golden_globes.film = golden_globes.film.str.strip()
golden_globes.nominee = golden_globes.nominee.str.strip()

#uppercase the film titles
golden_globes.film = golden_globes.film.str.upper()

#renaming columns
golden_globes = golden_globes.rename(columns={'category':'golden_globes_cat','win':'Globes_Winner'})
golden_globes

#renaming once upon a time in hollywood (adding a space after ...)
golden_globes.film = golden_globes.film.str.replace('ONCE UPON A TIME...IN HOLLYWOOD','ONCE UPON A TIME… IN HOLLYWOOD')
```

## Getting the Nominees for Golden Globes Best Director Category

```python
#creating a dataset with just directors

gg_directors = golden_globes.loc[golden_globes.golden_globes_cat=='Best Director - Motion Picture']
gg_directors = gg_directors.drop(columns=['year_film','ceremony','film'])
gg_directors
```

```python
#merging gg_directors and oscars

pictures_combined5 = pd.merge(pictures_combined4, gg_directors,
                             how='left',
                             left_on=['year_ceremony','Director'],
                             right_on=['year_award','nominee'])
```

```python
pictures_combined5 = pictures_combined5.rename(columns={'Globes_Winner':'Globes_Best_Dir_Winner'})
pictures_combined5 = pictures_combined5.drop(columns =['nominee'])
```

```python
pictures_combined5=pictures_combined5.drop_duplicates()
pictures_combined5
```

## Adding the nominees for Golden Globes Best Picture Category (incl. drama, comedy and musical)

```python
# Getting the list of films
gg_cats = ['Best Motion Picture - Drama','Best Motion Picture - Musical or Comedy','Picture','Picture - Comedy','Picture - Musical']


gg_pictures = golden_globes.loc[golden_globes.golden_globes_cat.isin(gg_cats)]

gg_pictures = gg_pictures.drop(columns=['film','year_film','ceremony'])
gg_pictures = gg_pictures.rename(columns={'Globes_Winner':'Globes_Best_Pic_Winner'})
gg_pictures.nominee =gg_pictures.nominee.str.upper()

#replacing Hollywood again
gg_pictures.nominee = gg_pictures.nominee.str.replace('ONCE UPON A TIME...IN HOLLYWOOD','ONCE UPON A TIME… IN HOLLYWOOD')
```

```python
pictures_combined5 = pd.merge(pictures_combined5, gg_pictures,
                             how='left',
                             left_on=['year_ceremony','film'],
                             right_on=['year_award','nominee'])
```

```python
pictures_combined5 = pictures_combined5.drop(columns = ['year_award_x','golden_globes_cat_x','year_award_y',
                                                       'golden_globes_cat_y','nominee'])
```

```python
pictures_combined5[['Globes_Best_Dir_Winner','Globes_Best_Pic_Winner']]= pictures_combined5[['Globes_Best_Dir_Winner','Globes_Best_Pic_Winner']].fillna('x')
```

```python
pictures_combined5.to_csv('data/pictures_combined5.csv', index=False)
```
